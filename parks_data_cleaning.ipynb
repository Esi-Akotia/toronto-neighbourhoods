```python
# Import dependencies

from pymongo import MongoClient
from config import mongo_esi_url
from pprint import pprint
import pandas as pd 
```

```python
# Call ddatabase collection from Mongo DB
mongo_uri = mongo_esi_url
client = MongoClient(mongo_esi_url)
db = client['opendata']
parks = db['parks']
```

```python
# Show contents of the database
park_results = parks.find()

# Iterate and print documents
for document in park_results:
    pprint(document)
```

```python
# Select relevant fields into lists
# Create empty lists
asset_id = []
name = []
asset_type = []
amenities = []

#Iterate to find relevant data
park_results = parks.find()
result = [x for x in park_results]
for item in result:
    asset_id.append(item['properties']['ASSET_ID'])
    name.append(item['properties']['ASSET_NAME'])
    asset_type.append(item['properties']['TYPE'])
    amenities.append(item['properties']['AMENITIES'])


# Put data into a dataframe
parks_df = pd.DataFrame({
    "ID": asset_id,
    "Name": name,
    "Type": asset_type,
    "Amenities": amenities 
})

# Formatting
parks_df.set_index("ID", inplace=True)
parks_df['Name'] = parks_df['Name'].str.title()

# Show values
parks_df.head(10)
```

```python
# Save to csv

parks_df.to_csv('clean_parks_data', index=False)
```
