```python
# Import dependencies

from pymongo import MongoClient
from config import mongo_esi_url
from pprint import pprint
import pandas as pd 
```

```python
# Call ddatabase collection from Mongo DB
mongo_uri = mongo_esi_url
client = MongoClient(mongo_esi_url)
db = client['opendata']
crime_data = db['crime']
```

```python
# Show contents of the database
crime_results = crime_data.find()

# Iterate and print documents
for document in crime_results:
    pprint(document)
```

```python
# Select relevant fields into lists
# Create empty lists
hood_id = []
area_name = []
population_2023 = []
assault_rate_2019 = []
assault_rate_2020 = []
assault_rate_2021 = []
assault_rate_2022 = []
assault_rate_2023 = []
autotheft_rate_2019 = []
autotheft_rate_2020 = []
autotheft_rate_2021 = []
autotheft_rate_2022 = []
autotheft_rate_2023 = []
biketheft_rate_2019 = []
biketheft_rate_2020 = []
biketheft_rate_2021 = []
biketheft_rate_2022 = []
biketheft_rate_2023 = []
breakenter_rate_2019 = []
breakenter_rate_2020 = []
breakenter_rate_2021 = []
breakenter_rate_2022 = []
breakenter_rate_2023 = []
homicide_rate_2019 = []
homicide_rate_2020 = []
homicide_rate_2021 = []
homicide_rate_2022 = []
homicide_rate_2023 = []
robbery_rate_2019 = []
robbery_rate_2020 = []
robbery_rate_2021 = []
robbery_rate_2022 = []
robbery_rate_2023 = []
shooting_rate_2019 = []
shooting_rate_2020 = []
shooting_rate_2021 = []
shooting_rate_2022 = []
shooting_rate_2023 = []
theftfrommv_rate_2019 = []
theftfrommv_rate_2020 = []
theftfrommv_rate_2021 = []
theftfrommv_rate_2022 = []
theftfrommv_rate_2023 = []
theftover_rate_2019 = []
theftover_rate_2020 = []
theftover_rate_2021 = []
theftover_rate_2022 = []
theftover_rate_2023 = []

#Iterate to find relevant data
crime_results = crime_data.find()
result = [x for x in crime_results]
for item in result:
    hood_id.append(item['properties']['HOOD_ID'])
    area_name.append(item['properties']['AREA_NAME'])
    population_2023.append(item['properties']['POPULATION_2023'])
    assault_rate_2019.append(item['properties']['ASSAULT_RATE_2019'])
    assault_rate_2020.append(item['properties']['ASSAULT_RATE_2020'])
    assault_rate_2021.append(item['properties']['ASSAULT_RATE_2021'])
    assault_rate_2022.append(item['properties']['ASSAULT_RATE_2022'])
    assault_rate_2023.append(item['properties']['ASSAULT_RATE_2023'])
    autotheft_rate_2019.append(item['properties']['AUTOTHEFT_RATE_2019'])
    autotheft_rate_2020.append(item['properties']['AUTOTHEFT_RATE_2020'])
    autotheft_rate_2021.append(item['properties']['AUTOTHEFT_RATE_2021'])
    autotheft_rate_2022.append(item['properties']['AUTOTHEFT_RATE_2022'])
    autotheft_rate_2023.append(item['properties']['AUTOTHEFT_RATE_2023'])
    biketheft_rate_2019.append(item['properties']['BIKETHEFT_RATE_2019'])
    biketheft_rate_2020.append(item['properties']['BIKETHEFT_RATE_2020'])
    biketheft_rate_2021.append(item['properties']['BIKETHEFT_RATE_2021'])
    biketheft_rate_2022.append(item['properties']['BIKETHEFT_RATE_2022'])
    biketheft_rate_2023.append(item['properties']['BIKETHEFT_RATE_2023'])
    breakenter_rate_2019.append(item['properties']['BREAKENTER_RATE_2019'])
    breakenter_rate_2020.append(item['properties']['BREAKENTER_RATE_2020'])
    breakenter_rate_2021.append(item['properties']['BREAKENTER_RATE_2021'])
    breakenter_rate_2022.append(item['properties']['BREAKENTER_RATE_2022'])
    breakenter_rate_2023.append(item['properties']['BREAKENTER_RATE_2023'])
    homicide_rate_2019.append(item['properties']['HOMICIDE_RATE_2019'])
    homicide_rate_2020.append(item['properties']['HOMICIDE_RATE_2020'])
    homicide_rate_2021.append(item['properties']['HOMICIDE_RATE_2021'])
    homicide_rate_2022.append(item['properties']['HOMICIDE_RATE_2022'])
    homicide_rate_2023.append(item['properties']['HOMICIDE_RATE_2023'])
    robbery_rate_2019.append(item['properties']['ROBBERY_RATE_2019'])
    robbery_rate_2020.append(item['properties']['ROBBERY_RATE_2020'])
    robbery_rate_2021.append(item['properties']['ROBBERY_RATE_2021'])
    robbery_rate_2022.append(item['properties']['ROBBERY_RATE_2022'])
    robbery_rate_2023.append(item['properties']['ROBBERY_RATE_2023'])
    shooting_rate_2019.append(item['properties']['SHOOTING_RATE_2019'])
    shooting_rate_2020.append(item['properties']['SHOOTING_RATE_2020'])
    shooting_rate_2021.append(item['properties']['SHOOTING_RATE_2021'])
    shooting_rate_2022.append(item['properties']['SHOOTING_RATE_2022'])
    shooting_rate_2023.append(item['properties']['SHOOTING_RATE_2023'])
    theftfrommv_rate_2019.append(item['properties']['THEFTFROMMV_RATE_2019'])
    theftfrommv_rate_2020.append(item['properties']['THEFTFROMMV_RATE_2020'])
    theftfrommv_rate_2021.append(item['properties']['THEFTFROMMV_RATE_2021'])
    theftfrommv_rate_2022.append(item['properties']['THEFTFROMMV_RATE_2022'])
    theftfrommv_rate_2023.append(item['properties']['THEFTFROMMV_RATE_2023'])
    theftover_rate_2019.append(item['properties']['THEFTOVER_RATE_2019'])
    theftover_rate_2020.append(item['properties']['THEFTOVER_RATE_2020'])
    theftover_rate_2021.append(item['properties']['THEFTOVER_RATE_2021'])
    theftover_rate_2022.append(item['properties']['THEFTOVER_RATE_2022'])
    theftover_rate_2023.append(item['properties']['THEFTOVER_RATE_2023'])

# Put data into a dataframe
crime_df = pd.DataFrame({
    'ID': hood_id,
    'Name': area_name,
    'Population': population_2023,
    'Assault rate 2019': assault_rate_2019,
    'Assault rate 2020': assault_rate_2020,
    'Assault rate 2021': assault_rate_2021,
    'Assault rate 2022': assault_rate_2022,
    'Assault rate 2023': assault_rate_2023,
    'Autotheft rate 2019': autotheft_rate_2019,
    'Autotheft rate 2020': autotheft_rate_2020,
    'Autotheft rate 2021': autotheft_rate_2021,
    'Autotheft rate 2022': autotheft_rate_2022,
    'Autotheft rate 2023': autotheft_rate_2023,
    'Biketheft rate 2019': biketheft_rate_2019,
    'Biketheft rate 2020': biketheft_rate_2020,
    'Biketheft rate 2021': biketheft_rate_2021,
    'Biketheft rate 2022': biketheft_rate_2022,
    'Biketheft rate 2023': biketheft_rate_2023,
    'Breakenter rate 2019': breakenter_rate_2019,
    'Breakenter rate 2020': breakenter_rate_2020,
    'Breakenter rate 2021': breakenter_rate_2021,
    'Breakenter rate 2022': breakenter_rate_2022,
    'Breakenter rate 2023': breakenter_rate_2023,
    'Homicide rate 2019': homicide_rate_2019,
    'Homicide rate 2020': homicide_rate_2020,
    'Homicide rate 2021': homicide_rate_2021,
    'Homicide rate 2022': homicide_rate_2022,
    'Homicide rate 2023': homicide_rate_2023,
    'Robbery rate 2019': robbery_rate_2019,
    'Robbery rate 2020': robbery_rate_2020,
    'Robbery rate 2021': robbery_rate_2021,
    'Robbery rate 2022': robbery_rate_2022,
    'Robbery rate 2023': robbery_rate_2023,
    'Shooting rate 2019': shooting_rate_2019,
    'Shooting rate 2020': shooting_rate_2020,
    'Shooting rate 2021': shooting_rate_2021,
    'Shooting rate 2022': shooting_rate_2022,
    'Shooting rate 2023': shooting_rate_2023,
    'Theft from motorvehicle rate 2019': theftfrommv_rate_2019,
    'Theft from motorvehicle rate 2020': theftfrommv_rate_2020,
    'Theft from motorvehicle rate 2021': theftfrommv_rate_2021,
    'Theft from motorvehicle rate 2022': theftfrommv_rate_2022,
    'Theft from motorvehicle rate 2023': theftfrommv_rate_2023,
    'Theft over $5,000 rate 2019': theftover_rate_2019,
    'Theft over $5,000 rate 2020': theftover_rate_2020,
    'Theft over $5,000 rate 2021': theftover_rate_2021,
    'Theft over $5,000 rate 2022': theftover_rate_2022,
    'Theft over $5,000 rate 2023': theftover_rate_2023
})

# formatting
crime_df.set_index("ID", inplace=True)

# Show values
crime_df.head(10)
```
