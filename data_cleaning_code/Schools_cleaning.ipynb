```python
import geopandas as gpd
```

```python
gdf = gpd.read_file('Resources/School locations-all types data - 4326.geojson')
gdf
```

```python
gdf.keys()
```

```python
gdf = gdf[['_id', 'NAME','BOARD_NAME','SOURCE_ADDRESS', 'SCHOOL_TYPE_DESC','ADDRESS_FULL', 'POSTAL_CODE',
       'MUNICIPALITY', 'CITY','geometry']]
gdf.set_index('_id', inplace=True)
```

```python
output_geojson_file = "schools_cleaned.geojson"
gdf.to_file(output_geojson_file, driver='GeoJSON')
```
